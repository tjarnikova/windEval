## get get from seasonal means

```python
%reset
import xarray as xr
import numpy as np
import glob
import matplotlib.pyplot as plt

sdir = '/gpfs/data/greenocean2/software/products/windsFromComponents/dailyStandard/'
idir = '/gpfs/data/greenocean2/software/products/windsFromComponents/dailyStandard/intProc/'

dss = ['UKESMEM']
#dss = []
def get_seasonal_jetlat(tdat, yrst=1980, yrend=2019):
    """Average winds to seasonal means first, then find jet latitude."""
    files = sorted(glob.glob(f'{sdir}{tdat}/{tdat}_wind_daily_1x1_*.nc'))
    files = [f for f in files if any(str(yr) in f for yr in range(yrst, yrend+1))]

    w = xr.open_mfdataset(files)
    q = w.sel(lat=slice(-70, -30))#.mean(dim='lon')

    seasons = {
        'FY':  None,
        'DJF': 'DJF',
        'MAM': 'MAM',
        'JJA': 'JJA',
        'SON': 'SON',
    }

    results = {}
    for sname, skey in seasons.items():
        if skey is None:
            smean = q.u10m.groupby('time_counter.year').mean(dim='time_counter')
        else:
            seasonal = q.u10m.sel(time_counter=(q.u10m['time_counter.season'] == skey))
            smean = seasonal.groupby('time_counter.year').mean(dim='time_counter')

        jetlat = smean.idxmax(dim='lat')
        results[sname] = jetlat

    return results


for ds in dss:
    print(f'Processing {ds}...')
    res = get_seasonal_jetlat(ds)
    for sname, jetlat in res.items():
        out = f'{idir}/{ds}_mnjet_seasfirstnolon_{sname}.nc'
        jetlat.to_netcdf(out)
        print(f'Saved {out}')


```

```python
def moving_average(timeseries, n = 3):
    # Ensure n is valid and doesn't exceed the length of the timeseries
    if n <= 0 or n > len(timeseries):
        raise ValueError("Window size n must be between 1 and the length of the timeseries.")
    
    # Compute the n-point moving average
    return np.convolve(timeseries, np.ones(n) / n, mode='valid')


import pymannkendall as mk
from scipy import stats

def give_trends(ts_y):

    ts_x = np.arange(0,len(ts_y))
    trend, h, mk_p, z, Tau, s, var_s, mk_slope, intercept = mk.original_test(ts_y)
    lin_slope, intercept, r_value, lin_p, std_err = stats.linregress(ts_x,ts_y)
    
    return mk_slope, mk_p, lin_slope, lin_p
```
